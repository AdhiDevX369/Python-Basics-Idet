## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=2,n_redundant=0)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
## Prediction
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
## Prediction
y_pred2=polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
## Prediction
y_pred3=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```

## SVM Decision Boundary Visualization

```python
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 7))
    plt.contourf(xx, yy, Z, levels=50, alpha=0.3, cmap='coolwarm')
    plt.contour(xx, yy, Z, levels=[0], colors='black', linewidths=2, linestyles='-')
    plt.contour(xx, yy, Z, levels=[-1], colors='blue', linewidths=1.5, linestyles='--', label='Margin -1')
    plt.contour(xx, yy, Z, levels=[1], colors='red', linewidths=1.5, linestyles='--', label='Margin +1')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', s=30)

    sv = model.support_vectors_
    plt.scatter(sv[:, 0], sv[:, 1], s=100, facecolors='none', edgecolors='yellow', linewidths=1.5, label='Support Vectors')

    mid_col = Z.shape[1] // 2
    col_vals = Z[:, mid_col]
    y_vals = yy[:, mid_col]
    x_pos = xx[0, mid_col]

    idx_zero = np.argmin(np.abs(col_vals))
    idx_neg = np.argmin(np.abs(col_vals + 1))
    idx_pos = np.argmin(np.abs(col_vals - 1))

    y_zero = y_vals[idx_zero]
    y_neg = y_vals[idx_neg]
    y_pos = y_vals[idx_pos]

    d = abs(y_pos - y_neg)

    plt.annotate('', xy=(x_pos, y_pos), xytext=(x_pos, y_neg),
                 arrowprops=dict(arrowstyle='<->', color='green', lw=2))
    plt.text(x_pos + 0.15, (y_pos + y_neg) / 2, f'd = {d:.2f}',
             fontsize=12, fontweight='bold', color='green',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

    plt.axhline(y=y_zero, color='black', linewidth=0.5, linestyle=':', alpha=0.5)

    plt.title(title, fontsize=14)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
plot_decision_boundary(svc, X_test, y_test, 'SVC - Linear Kernel')
```
