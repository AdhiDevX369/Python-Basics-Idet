## Introduction
Spam mail is a type of email that is sent to a large number of recipients without their consent. It is a type of electronic spam where unsolicited messages are sent by email. Many email spam messages are commercial in nature but may also contain disguised links that appear to be for familiar websites but in fact lead to phishing web sites or sites that are hosting malware. Spam email may also include malware as scripts or other executable file attachments. Definitions of spam usually include the aspects that email is unsolicited and sent in bulk.

## Problem Statement

The problem is to classify whether a mail is spam or not. The dataset contains 5572 rows and 2 columns. The first column is the text of the mail and the second column is the label of the mail. The label is 1 if the mail is spam and 0 if the mail is not spam.

The dataset contains 5572 rows and 2 columns. The first column is the text of the mail and the second column is the label of the mail. The label is 1 if the mail is spam and 0 if the mail is not spam.

# Importing Libraries

```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and Preprocessing

```python
# loading the data from the csv file using pandas
raw_mail_data = pd.read_csv('mail_data.csv')
raw_mail_data
```

```python
#replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
#print first 5 rows of data
mail_data.head()
```

```python
#checking the number of rows and columns in the dataframe
mail_data.shape
```

## Label Encoding

```python
# label spam mail as 0; and ham mail as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

--> Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering.

-- Spam Mail = 0

-- Not Spam Mail (Ham) = 1

```python
# seperating the data as text and label. X --> text; Y --> label
X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting the Data into Training and Testing Sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Future Extraction

Future extraction is a technique of extracting the features from the text data. It is a very important step in the text classification problem. The future extraction technique is used to convert the text data into numerical data.

```python
# transform the text data to feature vectors that can be used as input to the SVM model using TfidfVectorizer
future_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = future_extraction.fit_transform(X_train)
X_test_features = future_extraction.transform(X_test) 

# convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int') 
```

```python
print(X_train_features)
```

### TF-IDF Vectorizer

--> TF-IDF stands for Term Frequency-Inverse Document Frequency, is a numerical statistic that is intended to reflect how important a word is to a document in a collection or corpus.

--> TF-IDF is a combination of two terms. Term Frequency and Inverse Document Frequency.

--> Term Frequency is the ratio of the count of a word present in a sentence, to the length of the sentence.

--> Inverse Document Frequency is the logarithm of the ratio of the total number of sentences to the number of sentences containing the word.

### Training the Model

Logistic Regression
    --> Logistic Regression is a statistical method for analyzing a dataset in which there are one or more independent variables that determine an outcome. The outcome is measured with a dichotomous variable (in which there are only two possible outcomes).

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

Evaluation of the Trained Model

```python
# prediction on training data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# prediction on test data
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

## Building a Predictive System

System function to predict whether a mail is spam or not.

```python
#function
def predict_category(s):
    x = future_extraction.transform(s)
    y = model.predict(x)
    if y == [0]:
        return "This is a spam mail"
    else:
        return "This is a ham mail"


#predict the category using confusing mail 1
s1 = ["You have won! Click here to claim your reward and exclusive benefits today"]
print(predict_category(s1))

#predict the category using confusing mail 2
s2 = ["Hi, wanted to follow up on our discussion from yesterday about the project"]
print(predict_category(s2))
```

## Model Selection and Comparison


```python
# Create a dictionary of all models
models = {
    'Logistic Regression': model,
    'SVM': svm_model,
    'Naive Bayes': nb_model
}

# Create a comparison dataframe
comparison_data = {
    'Model': ['Logistic Regression', 'SVM', 'Naive Bayes'],
    'Training Accuracy': [accuracy_on_training_data, svm_accuracy_on_training_data, nb_accuracy_on_training_data],
    'Test Accuracy': [accuracy_on_test_data, svm_accuracy_on_test_data, nb_accuracy_on_test_data]
}

comparison_df = pd.DataFrame(comparison_data)
print("\n=== Model Comparison ===")
print(comparison_df)
```

```python
# Function to select model and make prediction
def predict_with_model(mail_text, model_name):
    if model_name not in models:
        return f"Error: Model '{model_name}' not found. Available models: {list(models.keys())}"
    
    selected_model = models[model_name]
    x = future_extraction.transform([mail_text])
    prediction = selected_model.predict(x)
    
    result = "Spam" if prediction == [0] else "Ham"
    return f"Model: {model_name} | Prediction: {result}"

# Test with different models
test_mail = "You have won! Click here to claim your reward and exclusive benefits today"
print(f"\nTesting: '{test_mail}'")
for model_name in models.keys():
    print(predict_with_model(test_mail, model_name))
```

## Model Extraction and Serialization


```python
import pickle
import os

# Create directory for saved models
model_dir = 'trained_models'
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

# Save all models
for model_name, model_obj in models.items():
    model_path = os.path.join(model_dir, f'{model_name.lower().replace(" ", "_")}.pkl')
    with open(model_path, 'wb') as f:
        pickle.dump(model_obj, f)
    print(f"Saved: {model_path}")

# Save the vectorizer
vectorizer_path = os.path.join(model_dir, 'vectorizer.pkl')
with open(vectorizer_path, 'wb') as f:
    pickle.dump(future_extraction, f)
print(f"Saved: {vectorizer_path}")
```

```python
# Function to load a saved model
def load_model(model_name):
    model_path = os.path.join(model_dir, f'{model_name.lower().replace(" ", "_")}.pkl')
    if os.path.exists(model_path):
        with open(model_path, 'rb') as f:
            loaded_model = pickle.load(f)
        print(f"Loaded model: {model_path}")
        return loaded_model
    else:
        print(f"Model not found: {model_path}")
        return None

# Function to load the vectorizer
def load_vectorizer():
    vectorizer_path = os.path.join(model_dir, 'vectorizer.pkl')
    if os.path.exists(vectorizer_path):
        with open(vectorizer_path, 'rb') as f:
            loaded_vectorizer = pickle.load(f)
        print(f"Loaded vectorizer: {vectorizer_path}")
        return loaded_vectorizer
    else:
        print(f"Vectorizer not found: {vectorizer_path}")
        return None

# Test loading a model
print("\n=== Testing Model Loading ===")
test_model = load_model('Logistic Regression')
test_vectorizer = load_vectorizer()
```

**Casual Conclusion**

- **Why Logistic Regression may have failed:** possible causes include class imbalance, insufficient hyperparameter tuning (regularization `C` / penalty), noisy or sparse TF‑IDF features, and features that need better preprocessing or feature selection. Logistic can also suffer when classes are not well separated in the current feature space.

- **Why standardization (and normalization) matters:** scaling (e.g., StandardScaler or L2 normalization) ensures features are on comparable scales, which helps gradient-based and distance-based models (Logistic, SVM) converge and learn meaningful weights. TF‑IDF vectors are partially normalized, but additional scaling or using L2 normalization consistently can still improve performance—especially when combining text features with other numeric features. Note: Naive Bayes typically does not require feature scaling.

- **Quick next steps:** run stratified cross-validation, tune hyperparameters (GridSearchCV or RandomizedSearchCV), try class weights or resampling if imbalance exists, and test standardizing/normalizing the input pipeline to see if Logistic/SVM improve.


## Hyperparameter Tuning — Grid Search & Randomized Search

We'll run GridSearchCV and RandomizedSearchCV for the three models, print best params and scores, then re-evaluate test/train accuracies and save the tuned models.

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import numpy as np

# Logistic Regression grid
lr_param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['liblinear', 'lbfgs']
}

# SVM random search space
svc_param_dist = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Naive Bayes grid
nb_param_grid = {
    'alpha': [0.1, 0.5, 1.0]
}

# Run GridSearch for LogisticRegression
print('\nRunning GridSearchCV for Logistic Regression...')
grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), lr_param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_lr.fit(X_train_features, Y_train)
print('Best LR params:', grid_lr.best_params_)
print('Best LR CV score:', grid_lr.best_score_)

# Run RandomizedSearch for SVM
print('\nRunning RandomizedSearchCV for SVM...')
rand_svc = RandomizedSearchCV(SVC(), svc_param_dist, n_iter=8, cv=5, scoring='accuracy', n_jobs=-1, random_state=42)
rand_svc.fit(X_train_features, Y_train)
print('Best SVM params:', rand_svc.best_params_)
print('Best SVM CV score:', rand_svc.best_score_)

# Run GridSearch for Naive Bayes
print('\nRunning GridSearchCV for Naive Bayes...')
grid_nb = GridSearchCV(MultinomialNB(), nb_param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_nb.fit(X_train_features, Y_train)
print('Best NB params:', grid_nb.best_params_)
print('Best NB CV score:', grid_nb.best_score_)

# Collect best estimators
best_lr = grid_lr.best_estimator_
best_svc = rand_svc.best_estimator_
best_nb = grid_nb.best_estimator_

# Update models dict to tuned versions
models_tuned = {
    'Logistic Regression (tuned)': best_lr,
    'SVM (tuned)': best_svc,
    'Naive Bayes (tuned)': best_nb
}

# Evaluate tuned models
tuned_results = []
for name, m in models_tuned.items():
    train_pred = m.predict(X_train_features)
    test_pred = m.predict(X_test_features)
    train_acc = accuracy_score(Y_train, train_pred)
    test_acc = accuracy_score(Y_test, test_pred)
    tuned_results.append((name, train_acc, test_acc))
    print(f"\n{name} -> Train acc: {train_acc:.4f}, Test acc: {test_acc:.4f}")

comparison_tuned_df = pd.DataFrame(tuned_results, columns=['Model', 'Training Accuracy', 'Test Accuracy'])
print('\n=== Tuned Model Comparison ===')
print(comparison_tuned_df)
```

```python
import pickle

if not os.path.exists(model_dir):
    os.makedirs(model_dir)

for name, m in models_tuned.items():
    safe_name = name.lower().replace(' ', '_').replace('(', '').replace(')', '')
    path = os.path.join(model_dir, f'{safe_name}.pkl')
    with open(path, 'wb') as f:
        pickle.dump(m, f)
    print(f"Saved tuned model: {path}")

with open(os.path.join(model_dir, 'vectorizer.pkl'), 'wb') as f:
    pickle.dump(future_extraction, f)
print('Saved/updated vectorizer.')

print('\nHyperparameter tuning and saving complete.')
```

```python
# Test the confusing mails with tuned models
confusing_mails = [
    "You have won! Click here to claim your reward and exclusive benefits today",
    "Hi, wanted to follow up on our discussion from yesterday about the project"
]

for mail in confusing_mails:
    print('\n---')
    print('Mail:', mail)
    for name, m in models_tuned.items():
        x = future_extraction.transform([mail])
        pred = m.predict(x)
        label = 'Spam' if int(pred[0]) == 0 else 'Ham'
        extra = ''
        if hasattr(m, 'predict_proba'):
            proba = m.predict_proba(x)[0]
            extra = f" | probs: {{spam:{proba[0]:.3f}, ham:{proba[-1]:.3f}}}"
        print(f"  {name}: {label} (raw: {pred}){extra}")

print('\nDone testing confusing mails with tuned models.')
```
